Kriteria Pengambilan Pemain: 
- Usia (Age) <= 25 tahun,
- Skill umum (Overall) >= 80 point, dan
- Potensi (Potential) >= 80 point.

```python
dfTarget = df[(df['Age'] <=25)&(df['Overall']>=80)&(df['Potential']>=80)]
print(len(dfTarget))
dfTarget.head(10)
```

Checking accuracy of data

#### Preparing for Model

```python
df['Recruit'] = df.apply(
lambda i: 1 if (i['Age'] <=25)&(i['Overall']>=80)&(i['Potential']>=80) else 0, axis=1)
```

```python
df.head()
df[['Age', 'Potential', 'Overall', 'Recruit']]
```

```python
len(df[df['Recruit'] == 1]) == len(dfTarget)
```

* Splitting


```python
from sklearn.model_selection import train_test_split

X = df[['Age','Overall','Potential']]
Y = df['Recruit']
```

```python
xtr, xts, ytr, yts = train_test_split(
    X,
    Y,
    test_size = .15
)
```

```python
print(X.loc[0])
```

```python
print(len(X))
print(len(Y))
print(len(xtr))
print(len(xts))
```

#### Fitting Data in Various Models

```python
from sklearn.linear_model import LogisticRegression
modelLR = LogisticRegression(solver='lbfgs', multi_class='auto')
modelLR.fit(xtr, ytr)
```

```python
from sklearn import tree
modelDT = tree.DecisionTreeClassifier()
modelDT.fit(xtr, ytr)
```

```python
from sklearn.ensemble import RandomForestClassifier

modelRF = RandomForestClassifier(n_estimators = 2)
modelRF.fit(xtr, ytr)
```

### Cross Validation Score

```python
from sklearn.model_selection import cross_val_score
```

```python
print('LogReg :', cross_val_score(
    LogisticRegression(solver='lbfgs', multi_class= 'auto'), xtr, ytr, cv=4
))

print('Random Forest :', cross_val_score(
    RandomForestClassifier(n_estimators = 5), xtr, ytr,
    cv=4
))

print('Decision Tree :', cross_val_score(
    tree.DecisionTreeClassifier(), xtr, ytr, cv=4
))

```

--> Akurasi terbaik dimiliki oleh Decision Tree dan Random Forest Models, namun ketiga model memiliki almost perfect accuracy. Mengantisipasi overfitting, diputuskan untuk menggunakan Random Forest
